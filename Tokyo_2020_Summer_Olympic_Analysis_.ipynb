```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = '/content/Tokyo 2020 Olympic.csv'
df = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataframe
df.head()
```

```python
# Basic information about the dataset
df.info()
```

```python
# Summary statistics
print(df.describe(include='all'))
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Unique values in each column
print("\nUnique values in each column:")
for column in df.columns:
    print(f"{column}: {df[column].nunique()} unique values")
```

```python
# Most frequent values in each column
print("\nMost frequent values in each column:")
for column in df.columns:
    print(f"{column}:")
    print(df[column].value_counts().head(10))
    print()
```

```python
# Number of medals by country
medals_by_country = df['country'].value_counts()
print("\nNumber of medals by country:")
print(medals_by_country.head(10))
```

```python
# Number of medals by discipline
medals_by_discipline = df['discipline'].value_counts()
print("\nNumber of medals by discipline:")
print(medals_by_discipline.head(10))
```

```python
# Number of medals by event
medals_by_event = df['event'].value_counts()
print("\nNumber of medals by event:")
print(medals_by_event.head(10))
```

```python
# Number of medals by athlete
medals_by_athlete = df['athlete_short_name'].value_counts()
print("\nNumber of medals by athlete:")
print(medals_by_athlete.head(10))
```

```python
# Number of medals by athlete's sex
medals_by_sex = df['athlete_sex'].value_counts()
print("\nNumber of medals by athlete's sex:")
print(medals_by_sex)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set the aesthetic style of the plots
sns.set_style("whitegrid")
```

```python
# Distribution of medals by type
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='medal_type', order=df['medal_type'].value_counts().index)
plt.title('Distribution of Medals by Type')
plt.xlabel('Medal Type')
plt.ylabel('Count')
plt.show()
```

```python
# Number of medals by country
plt.figure(figsize=(12, 8))
top_countries = df['country'].value_counts().nlargest(10).index
sns.countplot(data=df[df['country'].isin(top_countries)], x='country', hue='medal_type')
plt.title('Number of Medals by Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Medal Type')
plt.show()
```

```python
# Gender distribution of medal winners
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='athlete_sex', order=df['athlete_sex'].value_counts().index)
plt.title('Gender Distribution of Medal Winners')
plt.xlabel('Athlete Sex')
plt.ylabel('Count')
plt.show()
```

```python
# Number of medals by discipline
plt.figure(figsize=(12, 8))
top_disciplines = df['discipline'].value_counts().nlargest(10).index
sns.countplot(data=df[df['discipline'].isin(top_disciplines)], x='discipline', hue='medal_type')
plt.title('Number of Medals by Discipline')
plt.xlabel('Discipline')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Medal Type')
plt.show()
```

```python
# Medal distribution by event
plt.figure(figsize=(12, 8))
top_events = df['event'].value_counts().nlargest(10).index
sns.countplot(data=df[df['event'].isin(top_events)], x='event', hue='medal_type')
plt.title('Medal Distribution by Event')
plt.xlabel('Event')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Medal Type')
plt.show()
```

```python
# Number of medals by country and gender
plt.figure(figsize=(12, 8))
sns.countplot(data=df[df['country'].isin(top_countries)], x='country', hue='athlete_sex')
plt.title('Number of Medals by Country and Gender')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Athlete Sex')
plt.show()
```

```python
# Analysis of medal distribution by country and discipline
plt.figure(figsize=(12, 8))
sns.countplot(data=df[df['country'].isin(top_countries)], x='country', hue='discipline')
plt.title('Medal Distribution by Country and Discipline')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Discipline')
plt.show()
```

```python
# Gender distribution within top disciplines
plt.figure(figsize=(12, 8))
sns.countplot(data=df[df['discipline'].isin(top_disciplines)], x='discipline', hue='athlete_sex')
plt.title('Gender Distribution within Top Disciplines')
plt.xlabel('Discipline')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Athlete Sex')
plt.show()
```

```python
# Medals by gender and medal type
plt.figure(figsize=(10, 8))
sns.countplot(data=df, x='medal_type', hue='athlete_sex')
plt.title('Medals by Gender and Medal Type')
plt.xlabel('Medal Type')
plt.ylabel('Count')
plt.legend(title='Athlete Sex')
plt.show()
```

```python
# Athlete count per country
plt.figure(figsize=(14, 10))
athlete_count = df.groupby('country')['athlete_short_name'].nunique().nlargest(10).index
sns.countplot(data=df[df['country'].isin(athlete_count)], x='country')
plt.title('Athlete Count per Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count the number of medals won by each country
medals_by_country = df['country'].value_counts()
```

```python
# Get the top 5 highest medal-winning countries
top_5_countries = medals_by_country.head(5)
print("Top 5 highest medal-winning countries:")
print(top_5_countries)
```

```python
# Plot the top 5 highest medal-winning countries
plt.figure(figsize=(10, 6))
sns.barplot(x=top_5_countries.index, y=top_5_countries.values, palette='viridis')
plt.title('Top 5 Highest Medal-Winning Countries')
plt.xlabel('Country')
plt.ylabel('Number of Medals')
plt.show()
```

```python
# List of 10 different countries to display
countries = ['USA', 'China', 'Japan', 'Great Britain', 'Australia', 'Russia', 'Germany', 'France', 'Italy', 'Canada']

# Filter the dataframe to include only the specified countries
filtered_df = df[df['country'].isin(countries)]

# Count the number of medals won by each of the specified countries
medals_by_specified_countries = filtered_df['country'].value_counts()

# Plot the total medals won by the specified countries
plt.figure(figsize=(12, 8))
sns.barplot(x=medals_by_specified_countries.index, y=medals_by_specified_countries.values, palette='viridis')
plt.title('Total Medals Won by Specified Countries')
plt.xlabel('Country')
plt.ylabel('Number of Medals')
plt.xticks(rotation=45)
plt.show()

print("Total medals won by specified countries:")
print(medals_by_specified_countries)
```

```python
# Filter the dataframe to include only the medals won by India
india_medals = df[df['country'] == 'India']
```

```python
# Display the total number of medals won by India
total_medals_india = india_medals.shape[0]
print(f"Total number of medals won by India: {total_medals_india}")
```

```python
# Display the count of each type of medal won by India
medal_type_counts = india_medals['medal_type'].value_counts()
print("\nMedals won by India by type:")
print(medal_type_counts)
```

```python
# Plot the count of each type of medal won by India
plt.figure(figsize=(8, 6))
sns.countplot(data=india_medals, x='medal_type', order=medal_type_counts.index, palette='viridis')
plt.title('Medals Won by India by Type')
plt.xlabel('Medal Type')
plt.ylabel('Count')
plt.show()
```

```python
# Display medals by discipline
medals_by_discipline = india_medals['discipline'].value_counts()
print("\nMedals won by India by discipline:")
print(medals_by_discipline)
```

```python
# Plot medals by discipline
plt.figure(figsize=(12, 8))
sns.countplot(data=india_medals, x='discipline', order=medals_by_discipline.index, palette='viridis')
plt.title('Medals Won by India by Discipline')
plt.xlabel('Discipline')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Display medals by event
medals_by_event = india_medals['event'].value_counts()
print("\nMedals won by India by event:")
print(medals_by_event)
```

```python
# Plot medals by event
plt.figure(figsize=(12, 8))
sns.countplot(data=india_medals, x='event', order=medals_by_event.index, palette='viridis')
plt.title('Medals Won by India by Event')
plt.xlabel('Event')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
